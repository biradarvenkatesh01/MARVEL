```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
encoded_matrix = np.load('encoded_array.npy')
```

```python
total_elements = encoded_matrix.size
```

```python
side_length = int(np.sqrt(total_elements))
```

```python
squared_matrix = encoded_matrix.reshape(side_length, side_length)
```

```python
transposed_matrix = squared_matrix.T
```

```python
decoded_matrix = np.flipud(transposed_matrix)
```

```python
plt.imshow(decoded_matrix, cmap='gray')
plt.title("Decoded Image")
plt.show()
```

```python
flipped_axis0 = np.flip(decoded_matrix, axis=0)

plt.imshow(flipped_axis0, cmap='gray')
plt.title("Decoded Image")
plt.show()
```
